# Output2: Main code for the all review details


```python
import re, csv
Reviewdata = []
global check
check = 0

def getReviewData():
    global check
    allRview = driver.find_elements(By.CSS_SELECTOR, ".okeReviews-reviews-main .okeReviews-reviews-review")
    for review in range(check, len(allRview)):
        reviewer_name = allRview[review].find_element(By.CSS_SELECTOR, ".okeReviews-review-reviewer-profile-name").text
        rating = allRview[review].find_element(By.CSS_SELECTOR,".okeReviews-starRating .okeReviews-a11yText").text
        pattern = re.compile(r'\d+')
        rating= pattern.findall(rating)[0]
        review_body=allRview[review].find_element(By.CSS_SELECTOR," .okeReviews-review-main .okeReviews-review-main-heading").text
        date_of_review = allRview[review].find_element(By.CSS_SELECTOR,".okeReviews-review-date .okeReviews-a11yText + span").text
        Review_Description = allRview[review].find_element(By.CSS_SELECTOR, ".okeReviews-review-main .okeReviews-review-main-content-body").text
        try:
            verified_buyer_status=allRview[review].find_element(By.CSS_SELECTOR, ".okeReviews-review-reviewer-profile-details-inner + div").text
        except:
           verified_buyer_status = "Not Verifide"
        helpfuldata=allRview[review].find_element(By.CSS_SELECTOR, ".oke-helpful-vote").text
        # print(helpfuldata)
        HelpfullYes = pattern.findall(helpfuldata)[0]
        HelpfullNo = pattern.findall(helpfuldata)[1]
        
        data = {
            "reviewer name":reviewer_name,
            "rating": rating,
            "review title": review_body,
            "date of review": date_of_review,
            "Review body": Review_Description,
            "verified buyer status":verified_buyer_status,
            "HelpfullYes":HelpfullYes,
            "HelpfullNo":HelpfullNo,
        }
        Reviewdata.append(data)
    check = len(allRview)
    with open('reviews.csv', 'w', newline='', encoding='utf-8') as file:
        writer = csv.DictWriter(file, fieldnames=["reviewer name", "rating","review title", "date of review", "Review body", "verified buyer status","HelpfullYes","HelpfullNo"])
        writer.writeheader()
        for review_data in Reviewdata:
            print(review_data)
            writer.writerow(review_data)
```

```python
import time
while True:
    try:
        loadeMoreReviwes = driver.find_element(By.XPATH, "//button[normalize-space()='Show more']")
        loadeMoreReviwes.click()
        time.sleep(2)
        getReviewData()
    except Exception as e:
        print(e)
        break
```
